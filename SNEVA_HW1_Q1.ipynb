```python
from matplotlib import pylab as plt
import numpy as np
```

```python
#1a
#function for the binding energy, given values of A & Z
def bind_energy(A,Z):
    #constants
    a1= 15.8
    a2= 18.3
    a3= 0.714
    a4= 23.2

    #determining a5 based on the values of A & Z inputted
    if A %2 != 0:
        a5 = 0
    elif A %2 == 0 and Z %2 == 0:
        a5 = 12
    else:
        a5 = -12
    
    #calculation of the b_energy
    b_energy = ((a1*A) - (a2*A**(2/3)) - a3*((Z**2)/(A**(1/3))) - a4*(((A-2*Z)**2)/A) - (a5/(A**(1/2))))

    return b_energy

```

```python
#1b
#function for the energy per nucleon based on the binding energy
def nucleon_energy(A,Z):
    return bind_energy(A,Z)/A

A=58
Z=28
print('The binding energy is %f MeV.' % bind_energy(A,Z), 'The total binding energy per nucleon is %f MeV.' % nucleon_energy(A,Z))
```

```python
#1c
#function for any value of Z, finding the largest binding energy per nucleon.
def stable(Z):
    max_b_epn = 0.0
    max_A = 0
    
    #for loop for A going from Z to 3Z
    for A in range(Z, 3*Z+1):
        b_epn = nucleon_energy(A,Z)
        
        if max_b_epn < b_epn:             #finding the largest values for the binding energy per nucleon and A value
            max_b_epn = b_epn
            max_A = A
            
    return max_b_epn, max_A
```

```python
#input value of Z and calculate the maximum binding energy per nucleon

Z= int(input('select atomic number (Z):'))
max_b_epn, max_A = stable(Z)
print('For Z = %d: The maximum binding energy per nucleon = %f Mev for A = %d' % (Z, max_b_epn,max_A))
```

```python
#1d
#function for a sample of Z from 1 to 100 to find the most stable value for A of each one
def stable_sample():
    max_b_epn = 0
    max_Z = 0
    
    #for loop finding all of our A and bind energy per nucleon values
    for Z in range(1,101):
        max_stable, max_A = stable(Z)
        print('Z = %d: A = %d' % (Z,max_A))
        #binding energy and energy per nucleon
        b_energy = ((a1*A) - (a2*A**(2/3)) - a3*((Z**2)/(A**(1/3))) - a4*(((A-2*Z)**2)/A) - (a5/(A**(1/2))))
        b_epn = b_energy/A
        
        if max_b_epn < max_stable:       #determintes where the max binding energy occurs
            max_b_epn = max_stable
            max_Z = Z

    print('The maximum binding energy is at Z = %d.' % max_Z, 'The maximum binding energy per nucleon = %f Mev' % max_b_epn)
    return max_b_epn, max_Z
```

```python
#runs all samples of the function from Z=1 to Z=100. Returns the stable binding energy per nucleon and atomic number values
max_b_epn, max_A = stable_sample()
```

```python
fig= plt.figure()

ax= fig.add_subplot(1,1,1)
ax.set_title('Largest binding energy per Nucleon vs. Atomic Number')
ax.set_xlabel('Atomic Number')
ax.set_ylabel('Binding Energy per Nucleon')

plt.plot(stable_sample())
plt.show()
```
